```python
import numpy as np
import matplotlib
# matplotlib.rcParams['text.usetex'] = True
import matplotlib.pyplot as plt
# import covasim as cv
import math 



# Plotting a poisson distribution for my sanity

x = range(0,40) # x axis - number of events in timestep

lam = np.linspace(1,10,4) # lambda values to inspect

fact_x = np.zeros([np.size(x)]) # Factorial storage
P = np.zeros([np.size(lam),np.size(x)]) # Probability distributions

for ii in range(np.size(x)):
    fact_x[ii] = math.factorial(x[ii])

plt.figure()    # Plot distributions

for ii in range(np.size(lam)):
    for jj in range(np.size(x)):
        P[ii,jj] = lam[ii]**x[jj]*math.exp(-lam[ii])/fact_x[jj]

    plt.plot(x,P[ii,:],'-', label = ['lambda = ', round(lam[ii],2)])

plt.legend()
axes = plt.gca()
axes.set_xlim([0,20])
axes.set_ylim([0,0.5])
plt.show()
print('Sum of all distributions is: ', np.round(np.sum(P,1),1), '\nTherefore they are normalized' )

# Find how the peak of the distribution moves with lambda
x = range(0,40) # x axis - number of events in timestep

lam = np.linspace(1,10,20) # lambda values to inspect

Pmax = np.zeros([np.size(lam)])
Xmax = np.zeros([np.size(lam)])
P = np.zeros([np.size(lam),np.size(x)]) # Probability distributions
for ii in range(np.size(lam)):
    
    for jj in range(np.size(x)):
    
        P[ii,jj] = lam[ii]**x[jj]*math.exp(-lam[ii])/fact_x[jj]     
    
    Pmax[ii] = np.max(P[ii,:]) # Find maximum value
    
    Xmax[ii] = np.argmax(P[ii,:])+1 #  index of that value
   
    
    

plt.figure()    # Plot maximum probability point vs. lambda
plt.plot(lam,Xmax)
plt.xlabel('Lambda')
plt.ylabel('X_max')
plt.show()

plt.figure()  # Plot maximum probability point vs. lambda
plt.plot(lam,Pmax)
plt.xlabel('Lambda')
plt.ylabel('P_max')
plt.show()

# fig = plt.figure()
# # ax = fig.gca(projection='3d')
# plt.scatter(lam,Xmax,Pmax)
```
